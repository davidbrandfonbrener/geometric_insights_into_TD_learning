```python
f = np.load("../outputs/data/spiral/k_1_n_3_seed_1.npz")

tabular_Vs = f["tabular_Vs"]
spiral_Vs = f["spiral_Vs"]
condition_numbers = f["condition_numbers"]
bound = f["bound"]
smoothness = f["smoothness"]
deltas = f["deltas"]


n = deltas.shape[0]

colors = plt.cm.jet(np.linspace(0.2,0.8,n))
for i in range(0,spiral_Vs.shape[0]):
    plt.plot(spiral_Vs[i, :], color = colors[i], label = "%.2f" % np.sqrt(bound[i]))
plt.legend(title = "Reversibility")
plt.title("Varying delta")
plt.ylabel("Distance to V*")
plt.xlabel("Time")
plt.show()

colors = plt.cm.jet(np.linspace(0.2,0.8,n))
for i in range(0,spiral_Vs.shape[0]):
    plt.plot(tabular_Vs[i, :], color = colors[i], label = "%.2f" % bound[i])
plt.legend(title = "Reversibility")
plt.title("Tabular function")
plt.ylabel("Distance to V*")
plt.xlabel("Time")
plt.show()

tabular_Vs = []
f = np.load("../outputs/data/spiral/k_1_n_3_seed_1.npz")
spiral_Vs1 = f["spiral_Vs"]
tabular_Vs.append(f["tabular_Vs"])
f = np.load("../outputs/data/spiral/k_2_n_3_seed_1.npz")
spiral_Vs2 = f["spiral_Vs"]
tabular_Vs.append(f["tabular_Vs"])
f = np.load("../outputs/data/spiral/k_3_n_3_seed_1.npz")
spiral_Vs3 = f["spiral_Vs"]
tabular_Vs.append(f["tabular_Vs"])


n = deltas.shape[0]

colors = plt.cm.jet(np.linspace(0.2,0.8,5))

plt.plot(spiral_Vs1[0, :], color = colors[0], label = "1")
plt.plot(spiral_Vs2[0, :], color = colors[2], label = "2")
plt.plot(spiral_Vs3[0, :], color = colors[4], label = "3")


plt.legend(title = "k")
plt.title("Varying k")
plt.ylabel("Distance to V*")
plt.xlabel("Time")
plt.show()


colors = plt.cm.jet(np.linspace(0.2,0.8,n))
for i in range(0,len(tabular_Vs)):
    plt.plot(tabular_Vs[i][0,:], color = colors[i], label = "%.2f" % bound[i])
plt.legend(title = "Reversibility")
plt.title("Tabular function")
plt.ylabel("Distance to V*")
plt.xlabel("Time")
plt.show()
```
